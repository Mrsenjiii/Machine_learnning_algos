```python
def split_test_train() :
  np.random.seed(42) # this is for not changing the results
  shuffle_indx = np.random.permutation(len(data))
  test_size = int(len(data)*0.2)
  test_indx = shuffle_indx[:test_size]
  train_indx = shuffle_indx[test_size:]
  train,test=data.iloc[train_indx],data.iloc[test_indx]
  return (train,test)

train,test = split_test_train()
test.quality.hist()
```

```python
prop = data['quality'].value_counts() / len(data) # whole samplep proportion
prop_test=test['quality'].value_counts()/len(test) # test data proportion
fig,ax = plt.subplots(1,2 , sharey='row' , figsize =(10,5)) ##
# ax.bar(list(prop.index),list(prop.values))
ax[0].bar(list(prop.index),list(prop.values))
# fig1, ax1=plt.subplots()
ax[1].bar(list(prop_test.index),list(prop_test.values))
# print(prop.values)
# print(prop_test.values)
for x in prop_test.index:
  print(x , (prop.loc[x] - ((prop_test.loc[x]))/prop_test.loc[x])*100 , prop_test[x] , prop[x])
```

```python
# @title hellow
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index , test_index in split.split(data,data['quality']):
  train,test=data.loc[train_index].reset_index() ,data.loc[test_index].reset_index()


strata_ratio=test['quality'].value_counts()/len(test)
data_ratio= data['quality'].value_counts()/len(data)
```

```python
comparison=pd.DataFrame({'overall':data_ratio , 'strat_sampling':strata_ratio} )
comparison['comparison'] = (abs(comparison['overall'] - comparison['strat_sampling'])/comparison['overall'])*100
comparison
```
